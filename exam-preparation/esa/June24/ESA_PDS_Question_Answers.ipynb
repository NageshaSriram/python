Q1a. What is Dictionary Comprehension? Give an Example
Explanation: Dictionary comprehension is a concise way to create dictionaries in Python using a single line of code. It works similar to list comprehension but produces key–value pairs.

Example:

```python
# Create a dictionary of squares
squares = {x: x**2 for x in range(1, 6)}
print(squares)  
# Output: {1:1, 2:4, 3:9, 4:16, 5:25}
```

Q1c. What are map and reduce functions in Python?
Explanation:

map() applies a function to each element of an iterable.

reduce() (from functools) applies a function cumulatively to elements, reducing them to a single value.

Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]

# map: square each number
squares = list(map(lambda x: x**2, nums))  
# Output: [1, 4, 9, 16]

# reduce: sum of all numbers
total = reduce(lambda a, b: a+b, nums)     
# Output: 10
```

Q2c. Explain the difference between Pivot Table and Cross Table.
Explanation:

Pivot Table: Summarizes data, allows aggregation (mean, sum, count). Flexible.

Cross Table (crosstab): Shows frequency counts of combinations of two variables.

Example:

```python
df = pd.DataFrame({"Name":["Alice","Bob"],"Age":[25,20]})

# Pivot table
pd.pivot_table(df, values="Age", index="Name", aggfunc="mean")

# Crosstab
pd.crosstab(df["Name"], df["Age"])
```

### 3a.)You are working with NASA and suddenly you got an opportunity to travel to the moon. You have to stay there for 15 years.
 You have to calculate your weight in kilograms on the moon in a period of 15 years
 Step1: Measure your Weight on the Moon and also create a variable to store it. Assume your starting weight:30
 For each year, you can calculate the new weight by adding a kilogram, and then multiplying by 6.5 percent (0.65) to get the weight on the moon  (5 Marks)


Q3a. Weight on the Moon over 15 years
Explanation: Starting weight = 30 kg. Each year: add 1 kg, then multiply by 0.65 (since Moon gravity is 65% of Earth’s). We store yearly weights in a list.

```python
weight_earth = 30
moon_weights = []

for year in range(1, 16):
    weight_earth += 1
    weight_moon = weight_earth * 0.65
    moon_weights.append((year, weight_moon))

print("Yearly Moon Weights:", moon_weights)
```

Output (sample): Year 1 → 20.15 kg, Year 2 → 20.8 kg … Year 15 → ~28.6 kg.

    List = [12,6,4,7,4,7,67,88,99,355,7887,6,467,88,4,77,8]? (5 Marks)

Q3d. Remove duplicates while preserving order
Explanation: Use a set to track seen elements.

Code:

```python
given_list = [12,6,4,7,4,7,67,88,99,355,7887,6,467,88,4,77,8]
result = []
seen = set()

for item in given_list:
    if item not in seen:
        result.append(item)
        seen.add(item)

print("List without duplicates:", result)
```

 ### 3e.) Given two list for names of the users and their birthdays respectively.
Write a program to answer following:

a) Print out the name of the youngest user

b) Print out the names of the users born in Leap year.

Names = ['Alice', 'Bob', 'Caroline','Derek','Edwin']

Birthdays = ['4/7/1990', '11/12/1995', '8/17/2000','9/4/2004','2/29/2000'](5 Marks)

Explanation: Parse birthdays, find youngest (latest year), and check leap years.

Code:

```python
from datetime import datetime

names = ['Alice', 'Bob', 'Caroline','Derek','Edwin']
birthdays = ['4/7/1990', '11/12/1995', '8/17/2000','9/4/2004','2/29/2000']

# Convert to datetime
dates = [datetime.strptime(b, "%m/%d/%Y") for b in birthdays]

# Youngest user
youngest_index = dates.index(max(dates))
print("Youngest user:", names[youngest_index])

# Leap year check
leap_year_names = []
for name, date in zip(names, dates):
    year = date.year
    if (year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)):
        leap_year_names.append(name)

print("Users born in leap year:", leap_year_names)
```

#### 3f.) A superstore has multiple items to sell. These items are categorised as  category A and category B. For festival time the superstore decided to declare some discount, as given below. 
For category A items, it charges Rs.15 per item if you buy less than 30 quantity of an item. If you buy between 30 and 99 quantity of an item, the cost is Rs.12 per item. If buying quantity is 100 or more for an item, the cost is Rs.10 per item. 

For category B items, it charges Rs.30 per item if you buy less than 50 quantity of an item. If you buy between 50 and 99 quantity of an item, the cost is Rs.25 per item. If buying quantity is 100 or more for an item, the cost is Rs.20 per item. Further it also gives discount of 5% if total bill exceeds 1000 and 10% if bill exceeds 1500.

Write a program which repeatedly ask user to enter no of items and its category. If user enters 'STOP' it shows user has done with the purchase, now calculate total payable amount and display it to user(10 Marks)


***
 Sample Input/Output:
Enter no of items and its category: 23,A

Enter no of items and its category: 56, b

Enter no of items and its category: 67,b

Enter no of items and its category: 110,a

Enter no of items and its category: STOP

Total amount payable is :  2808.0

***

Q3f. Superstore Discount Program
Explanation: Implement rules for category A and B, apply discounts if bill exceeds thresholds.

```python
total = 0

while True:
    entry = input("Enter no of items and its category: ")
    if entry.upper() == "STOP":
        break
    qty, cat = entry.split(",")
    qty = int(qty)
    cat = cat.strip().upper()

    if cat == "A":
        if qty < 30: cost = qty * 15
        elif qty < 100: cost = qty * 12
        else: cost = qty * 10
    elif cat == "B":
        if qty < 50: cost = qty * 30
        elif qty < 100: cost = qty * 25
        else: cost = qty * 20
    total += cost

# Apply discount
if total > 1500:
    total *= 0.90
elif total > 1000:
    total *= 0.95

print("Total amount payable is:", total)
```

1. Find the number of recovered and Hospitalized cases in Maharashtra'? (4 Marks)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("covid.csv")
df.head()
```

### 1. Find the number of recovered and Hospitalized cases in Maharashtra'? (4 Marks)


```python
# Explanation: We filter rows where Detected State == 'Maharashtra' and then count values in the Current Status column.

# Filter Maharashtra cases
mh_cases = df[df['Detected State'] == 'Maharashtra']

# Count by Current Status
status_counts = mh_cases['Current Status'].value_counts()
print(status_counts)

# Output (from your dataset snippet):

# Hospitalized: many rows (majority)

# Recovered: some rows (e.g., 18146–18149, 18175–18179 etc.)
```

###  2. Find the number of Female and Male Hospitalized  from Maharashtra'? (4 Marks)

```python
# Explanation: Filter Maharashtra + Hospitalized, then group by Gender.

mh_hosp = mh_cases[mh_cases['Current Status'] == 'Hospitalized']
gender_counts = mh_hosp['Gender'].value_counts()
print(gender_counts)

# Output (from snippet):

# Males: higher count

# Females: fewer rows (e.g., 18110, 18119, 18120, etc.)
```

### 3. How many people Travelled from Wuhan in Maharastra. (4 Marks)






```python
# Explanation: Look at Notes column for “Travelled from Wuhan”.

wuhan_cases = mh_cases[mh_cases['Notes'].str.contains("Wuhan", na=False)]
print(len(wuhan_cases))

# Output (from snippet): At least cases 15229, 15230, 18216, 18217 → 4 cases.
```

### 4. Find the percentage of male and female got COVID-19 using pie chart (5 Marks)

```python
# Explanation: Count gender across all patients, then plot pie chart.

import matplotlib.pyplot as plt

gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title("Percentage of Male vs Female COVID-19 Patients")
plt.show()

# Interpretation: Pie chart shows distribution. In your dataset, male patients dominate.
```

### 5 .Find the state with the highest number of female patients and the state with the highest number of male patients.(5 marks)

```python
# Explanation: Group by state and gender, then find max.

gender_state = df.groupby(['Detected State','Gender']).size().reset_index(name='count')

# Highest female
female_max = gender_state[gender_state['Gender']=='F'].sort_values('count', ascending=False).head(1)

# Highest male
male_max = gender_state[gender_state['Gender']=='M'].sort_values('count', ascending=False).head(1)

print("Highest Female Patients:", female_max)
print("Highest Male Patients:", male_max)

# Output (from snippet): Likely Maharashtra has both highest male and female counts.
```

### 6. Display the count plot to show the number of people got affected based on states. (5 Marks)

```python
# Explanation: Use seaborn countplot.

import seaborn as sns

plt.figure(figsize=(12,6))
sns.countplot(y='Detected State', data=df, order=df['Detected State'].value_counts().index)
plt.title("Number of COVID-19 Cases by State")
plt.show()

# Interpretation: Maharashtra and Uttar Pradesh dominate in your dataset.
```

### 7. Find the number of cases in each state, categorized by age group (Child, Adult, Senior).(8 marks)

```python
# Explanation: Define age groups:

# Child: <18

# Adult: 18–59

# Senior: ≥60 Then group by state + age group.

def age_group(age):
    try:
        age = int(age)
        if age < 18: return 'Child'
        elif age < 60: return 'Adult'
        else: return 'Senior'
    except:
        return 'Unknown'

df['AgeGroup'] = df['Age Bracket'].apply(age_group)
age_counts = df.groupby(['Detected State','AgeGroup']).size().reset_index(name='count')
print(age_counts)

# Output: Most rows have missing age, but where present (e.g., 20, 24, 45, 69), they fall into Adult or Senior.
```

### 8.Calculate the percentage of patients who recovered in Telangana compared to the total number of cases in Telangana.(5 marks)

```python
# Explanation: Filter Telangana cases, then compute recovered / total.

tg_cases = df[df['Detected State'] == 'Telangana']
recovered_tg = tg_cases[tg_cases['Current Status'] == 'Recovered']
percentage = (len(recovered_tg) / len(tg_cases)) * 100
print("Recovered % in Telangana:", percentage)

# Output (from snippet): Telangana has 1 recovered case (TS-P1) out of total Telangana cases → 100%.
```
