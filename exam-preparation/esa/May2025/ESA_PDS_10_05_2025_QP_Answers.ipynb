Q1 (a) How are dictionaries different from lists in Python?

Lists: Ordered collections of elements, accessed by index (0,1,2â€¦).

Dictionaries: Unordered collections of key-value pairs, accessed by keys.

```python
# List
my_list = [10, 20, 30]
print(my_list[1])   # Output: 20

# Dictionary
my_dict = {"name": "Nagesha", "age": 25}
print(my_dict["name"])  # Output: Nagesha

# Difference: Lists use positions, dictionaries use keys.
```

Q1 (b) What is a lambda function? How is it different from a regular function?

Lambda function: Anonymous, single-line function defined using lambda.

Regular function: Defined using def, can have multiple statements.

Example:

```python
# Regular function
def square(x):
    return x*x

# Lambda function
square_lambda = lambda x: x*x

print(square(5))        # Output: 25
print(square_lambda(5)) # Output: 25

# ðŸ‘‰ Difference: Lambda is concise, used for short operations; regular functions are more flexible.
```

Q2 (e) Explain the difference between pivot table and cross table.

- Pivot Table:

Summarizes data by grouping and aggregating.

Flexible: can calculate sums, averages, counts.

Example: Pandas pivot_table().

- Cross Table (Crosstab):

Shows frequency counts of combinations of two variables.

Example: Pandas crosstab().

Example:

```python
import pandas as pd

data = {'Gender':['M','F','M','F'], 'Course':['CS','CS','AI','AI'], 'Marks':[80,90,70,85]}
df = pd.DataFrame(data)

# Pivot table
print(pd.pivot_table(df, values='Marks', index='Gender', columns='Course', aggfunc='mean'))

# Crosstab
print(pd.crosstab(df['Gender'], df['Course']))

# ðŸ‘‰ Pivot â†’ aggregation (mean, sum). Crosstab â†’ frequency counts.
```

 ### Q3a. Write a Python program that takes a list of integers as input and returns a new list containing only the prime numbers that are also palindromes (e.g., 131, 7, 11). If no such numbers are found, return an empty list. (8 Marks)

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def is_palindrome(n):
    return str(n) == str(n)[::-1]

def prime_palindromes(nums):
    return [num for num in nums if is_prime(num) and is_palindrome(num)]

# Example
nums = [7, 11, 121, 131, 23, 101]
print(prime_palindromes(nums))  # Output: [7, 11, 131, 101]
```

Explanation:

is_prime() checks primality.

is_palindrome() checks if number reads same forward/backward.

List comprehension filters numbers satisfying both.

### Q3b. Write a Python program to manage a basic student marks system. Your program should:(8 Marks)

1.Allow the user to add a new student with their name and marks.

2.Implement a function to update the marks of an existing student.

3.Implement a function to calculate the average marks of all students.

4.Implement a function to display all student names with their marks.


```python
students = {}

def add_student(name, marks):
    students[name] = marks

def update_marks(name, marks):
    if name in students:
        students[name] = marks
    else:
        print("Student not found!")

def average_marks():
    return sum(students.values()) / len(students) if students else 0

def display_students():
    for name, marks in students.items():
        print(f"{name}: {marks}")

# Example usage
add_student("Alice", 85)
add_student("Bob", 90)
update_marks("Alice", 95)
print("Average:", average_marks())
display_students()
```

Explanation:

Dictionary stores student â†’ marks.

Functions handle CRUD operations.

Average calculated using sum()/len().

 ### Q3c. Create a Python function that takes a dictionary of employee names with their corresponding (years of experience, base bonus, and performance rating on a scale of 1â€“5). The function should calculate and return a new dictionary with adjusted bonuses based on the following rules:(8 Marks)

If an employee has more than 10 years of experience, increase their bonus by 20% plus 3% of the base bonus multiplied by their performance rating.

If they have between 5 and 10 years of experience, increase their bonus by 15% plus 2% of the base bonus multiplied by their performance rating.

If they have less than 5 years of experience, increase their bonus by 10% plus 1% of the base bonus multiplied by their performance rating.

The function should return a new dictionary with employee names and their updated bonus amounts.

Example:

Input:

employees = {
    "Raju": (12, 3000, 4),
    "Somu": (8, 2500, 3),
    "Ravi": (3, 2000, 5),
    "David": (15, 4000, 2)
}

Output:

Adjusted salaries:
Raju: $3960.0,
Somu: $3025.0,
Ravi: $2300.0,
David: $5040.0


Problem: Adjust bonuses based on experience and performance rating.

Code:

```python
def calculate_bonus(employees):
    updated = {}
    for name, (years, base_bonus, rating) in employees.items():
        if years > 10:
            bonus = base_bonus * (1.20 + 0.03 * rating)
        elif 5 <= years <= 10:
            bonus = base_bonus * (1.15 + 0.02 * rating)
        else:
            bonus = base_bonus * (1.10 + 0.01 * rating)
        updated[name] = round(bonus, 2)
    return updated

# Example
employees = {
    "John": (12, 10000, 4),
    "Mary": (7, 8000, 5),
    "Sam": (3, 5000, 2)
}
print(calculate_bonus(employees))
```

Explanation:

Rules applied with conditional logic.

Multipliers adjust base bonus.

Rounded for neatness.

 ### Q3d. Write a Python program that takes a list of sentences and returns a dictionary where the keys are the unique words (case insensitive) that are at least 4 characters long, and the values are the counts of those words. Ignore punctuation and consider only alphabetic characters. (8 marks)


Example:

Input:

sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Is this the real life? It is just fantastic!",
    "And in the end, the love you take is equal to the hate you make.",
    "To be or not to be, that is the question."
]

Output:

Word frequency (length >= 4): {'quick': 1, 'brown': 1, 'jumps': 1, 'over': 1, 'lazy': 1, 'this': 1, 'real': 1, 'life': 1, 'just': 1, 'fantastic': 1, 'love': 1, 'take': 1, 'equal': 1, 'hate': 1, 'make': 1, 'that': 1, 'question': 1}

Problem: From sentences, count unique words â‰¥4 characters, case-insensitive, ignoring punctuation.

```python
import re

def word_count(sentences):
    word_dict = {}
    for sentence in sentences:
        words = re.findall(r"[a-zA-Z]+", sentence.lower())
        for word in words:
            if len(word) >= 4:
                word_dict[word] = word_dict.get(word, 0) + 1
    return word_dict

# Example
sentences = ["Python is great!", "Programming in Python is fun.", "Fun with coding."]
print(word_count(sentences))
```

Explanation:

Regex extracts alphabetic words.

Converted to lowercase for case-insensitivity.

Dictionary counts words â‰¥4 characters.

 ### Q3e. Given two lists containing the names of students and their respective dates of birth (in the format "YYYY-MM-DD"), write a Python program to answer the following:

a) Print the name of the student who is the youngest. (4 marks)

b) Print the names of students who were born in the month of December. (4 marks)

note:
names = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
birthdays = ["2005-06-15", "2003-12-22", "2004-05-10", "2006-12-05", "2003-01-11"]


Q3 (e) Studentsâ€™ DOB Analysis
Problem: Given names + DOBs, find youngest student and those born in December.

Code:

```python
from datetime import datetime

def analyze_dobs(names, dobs):
    dob_dict = {name: datetime.strptime(dob, "%Y-%m-%d") for name, dob in zip(names, dobs)}
    
    # Youngest student
    youngest = max(dob_dict, key=dob_dict.get)
    print("Youngest student:", youngest)
    
    # Students born in December
    december_students = [name for name, dob in dob_dict.items() if dob.month == 12]
    print("Born in December:", december_students)

# Example
names = ["Alice", "Bob", "Charlie"]
dobs = ["2000-12-15", "1999-05-20", "2001-12-01"]
analyze_dobs(names, dobs)
```

Explanation:

datetime.strptime parses DOBs.

max() finds latest date â†’ youngest.

Month check (dob.month == 12) filters December births.

Q4 (a) Car Pricing & Mileage Analysis
(i) Cars with price > 25,000 and average mileage

```python
import pandas as pd

df = pd.read_csv("car_info.csv")

# Cars with price > 25,000
high_price_cars = df[df["Price"] > 25000]
count_high_price = len(high_price_cars)
avg_mileage_high_price = high_price_cars["Mileage"].mean()

print("Number of cars with price > 25,000:", count_high_price)
print("Average mileage of these cars:", avg_mileage_high_price)
```

Inference:

This tells us how many premium cars are in inventory.

Average mileage helps understand whether high-priced cars are newer (low mileage) or older (high mileage).

```python
# (ii) Average price by fuel type
avg_price_by_fuel = df.groupby("Fuel_Type")["Price"].mean()
print(avg_price_by_fuel)
```

Inference:

Compare Petrol vs Diesel average prices.

Useful for dealership pricing strategy.

```python
# (iii) Inventory summary by make

summary = df.groupby("Make").agg(
    avg_price=("Price", "mean"),
    max_mileage=("Mileage", "max")
)
print(summary)
```

Inference:

Shows brand-level performance.

Helps identify which brands have higher mileage cars (older stock).

```python
# (iv) Mileage Category classification

def mileage_category(m):
    if m < 30000:
        return "Low"
    elif m < 100000:
        return "Moderate"
    else:
        return "High"

df["Mileage_Category"] = df["Mileage"].apply(mileage_category)
print(df[["Make","Mileage","Mileage_Category"]].head())
```

Inference:

Categorizes cars into usage levels.

Useful for customer segmentation (new vs used buyers).

```python
# (v) Highest-priced car per make
highest_price_per_make = df.groupby("Make").apply(lambda x: x.loc[x["Price"].idxmax()])
print(highest_price_per_make[["Make","Model","Price"]])
```
