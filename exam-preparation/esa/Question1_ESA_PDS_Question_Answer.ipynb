### 2d. What are map and reduce functions in Python?
- map: Applies a function to every item in an iterable (like a list) and returns a map object (can be converted to a list).

python
result = list(map(lambda x: x*2, [1, 2, 3]))  # [2, 4, 6]

- reduce: Repeatedly applies a binary function to the items of an iterable, reducing it to a single cumulative value. It is in the functools module.

python
from functools import reduce
sum_val = reduce(lambda x, y: x + y, [1, 2, 3])  # 6
map is for transforming data elements; reduce is for aggregating data to a single result.​

```python
# 2a. How to merge dataframes in pandas?

import pandas as pd

def merge_dataframes(df1, df2, on_column, how='inner'):
    """
    Merge two dataframes on a given column using specified method.
    
    Parameters:
        df1 (pd.DataFrame): First dataframe
        df2 (pd.DataFrame): Second dataframe
        on_column (str): Column name to merge on
        how (str): Type of merge - 'inner', 'outer', 'left', 'right'
    
    Returns:
        pd.DataFrame: Merged dataframe
    """
    merged_df = pd.merge(df1, df2, on=on_column, how=how)
    return merged_df


# 2b. Define significant features of Numpy library.

numpy_features = """
- Provides support for large, multi-dimensional arrays and matrices.
- Offers a large collection of high-level mathematical functions to operate on these arrays.
- Enables efficient numerical computation with broadcasting, vectorization, and integrated C and Fortran code.
"""

# 2c. How to rank the rows in pandas dataframe based on a specific column?

def rank_dataframe(df, rank_column, ascending=True):
    """
    Rank rows of a dataframe based on specified column.
    
    Parameters:
        df (pd.DataFrame): Input dataframe
        rank_column (str): Column based on which ranking is to be done
        ascending (bool): Rank in ascending order if True, else descending
    
    Returns:
        pd.DataFrame: Dataframe with an added 'Rank' column
    """
    df['Rank'] = df[rank_column].rank(ascending=ascending)
    return df


# 2d. What are different types of joins in pandas?

join_types = """
- inner: Returns intersection of keys from both dataframes.
- outer: Returns union of keys, with NaN for missing matches.
- left: Returns keys from left dataframe with matched keys from right.
- right: Returns keys from right dataframe with matched keys from left.
"""

# 2e. Basic syntax of reading a csv file in pandas and_display the first 5 records

def read_csv_file(file_path):
    """
    Read a CSV file and return the first 5 rows.
    
    Parameters:
        file_path (str): Path to CSV file
    
    Returns:
        pd.DataFrame: First 5 rows of the CSV file
    """
    df = pd.read_csv(file_path)
    return df.head()

```

### Question 3a.i

i) You are working on a project to develop a simple online shopping application. You need to
implement various data structures in Python to manage the shopping cart, product inventory, and user information efficiently.

You have been given a list of products with their names and prices. Write a Python function named calculate_total_cost that takes a list of products and their corresponding quantities in the shopping cart and returns the total cost of the items.

```python
def calculate_total_cost(products, quantities):
    """
    Calculate the total cost of items in the shopping cart.
    
    Parameters:
    products (list of tuples): List containing tuples of (product_name, price)
    quantities (list of int): List containing corresponding quantities for each product
    
    Returns:
    float: Total cost of all products with quantities
    """
    total_cost = 0
    for (product, price), quantity in zip(products, quantities):
        total_cost += price * quantity
    return total_cost

# Example Usage
products = [('Laptop', 50000), ('Mouse', 500), ('USB Cable', 200)]
quantities = [1, 2, 3]

total = calculate_total_cost(products, quantities)
print("Total Cost:", total)
```

### 3(a)(iii) Update inventory

```python
def update_inventory(product_inventory, new_shipment):
    for product, qty in new_shipment.items():
        product_inventory[product] = product_inventory.get(product, 0) + qty
    return product_inventory

shipment = {"Laptop": 5, "Monitor": 10}
print(update_inventory(product_inventory, shipment))
```

### 3(a)(iv) Apply discount

```python
def apply_discount(product_prices, discount_dict):
    for product, new_price in discount_dict.items():
        if product in product_prices:
            product_prices[product] = new_price
    return product_prices

prices = {"Laptop": 50000, "Mouse": 500}
discounts = {"Laptop": 45000}
print(apply_discount(prices, discounts))  # {'Laptop': 45000, 'Mouse': 500}
```

### 3(b) Scientific Calculator

```python
import math

def calculator():
    print("Choose operation: 1.Arithmetic 2.Trigonometric 3.Logarithmic")
    choice = int(input("Enter choice: "))
    
    if choice == 1:
        a, b = map(float, input("Enter two numbers: ").split())
        print("Sum:", a+b, "Diff:", a-b, "Prod:", a*b, "Div:", a/b)
    elif choice == 2:
        angle = float(input("Enter angle in radians: "))
        print("sin:", math.sin(angle), "cos:", math.cos(angle), "tan:", math.tan(angle))
    elif choice == 3:
        num = float(input("Enter number: "))
        print("log:", math.log(num), "log10:", math.log10(num))

calculator()
```

### 3(c) Reverse words

```python
def reverse_words(sentence):
    return " ".join(sentence.split()[::-1])

print(reverse_words("Python for Data Science"))  # "Science Data for Python"
```

### 3(d) Find special codes

```python
def find_special_codes(message):
    words = message.split()
    return [w for w in words if w.isalnum()]

msg = "abc123 @#$ xyz789"
print(find_special_codes(msg))  # ['abc123', 'xyz789']
```

### 4(b) Athlete Events + Regions dataset

```python
athletes = pd.read_csv("athlete_events.csv")
regions = pd.read_csv("regions.csv")

# 1. Merge
merged = athletes.merge(regions, on="NOC", how="left")

# 2. Age distribution of gold medalists
gold = merged[(merged['Medal']=='Gold')]
gold['Age'].hist()

# 3. Gold medalists age > 50
print(gold[gold['Age']>50])

# 4. Women who played in Summer
print(merged[(merged['Sex']=='F') & (merged['Season']=='Summer')])

# 5. Top 5 countries with most medals
medals = merged.groupby('region')['Medal'].count().sort_values(ascending=False).head(5)
print(medals)
medals.plot(kind='bar')
```
