```python
# Explain the difference between pivot table and crosstab in pandas.
# A pivot table is a data summarization tool that allows you to aggregate and reshape data based on specified index and column values, often used for multi-dimensional analysis. 
# A crosstab, on the other hand, is a specific type of pivot table that is used to compute a simple frequency table of two or more categorical variables, typically used for contingency tables.

# Example for pivot table and crosstab
import pandas as pd 
data = {
    'A': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'],
    'B': ['one', 'one', 'two', 'two', 'one', 'one'],
    'C': [1, 2, 3, 4, 5, 6],
    'D': [10, 20, 30, 40, 50, 60]
} 
df = pd.DataFrame(data)
pivot_table = pd.pivot_table(df, values='D', index=['A'], columns=['B'], aggfunc='sum')
crosstab = pd.crosstab(df['A'], df['B'])  
print("Pivot Table:\n", pivot_table)
print("Crosstab:\n", crosstab)
```

3.a. James Deen is an investor specializing in the Technology Domain. James purchases stock in successful technology companies using a stockbroker and later sells them at a profit margin

As part of his investment portfolio, James purchased 2000 units of stock in Google LLC worth $80,000 ($40 apiece) in the month of November, 2019. The stockbroker charged 3% commission on the total purchase value of the stocks.A month later, James sells his entire stock holding in Google LLC for $42.75 apiece. He pays a commission of 3% on this transaction as well to his stockbroker.

**Write a Python program that returns the following output:**

1. The cost of purchasing the entire stock holding in Google LLC

2. The amount of money paid in commission during the purchase

3. The selling price of the entire stock holding in Google LLC

4. The amount of money paid in commission during the sale

5. The profit margin made by James in the entire transaction (purchase & sale, including commissions paid)

```python
(80000 + 2400.0) - (85500.0 - 2565.0)
```

```python
# Given data
units = 2000
purchase_price_per_stock = 40
selling_price_per_stock = 42.75
commission_rate = 0.03  # 3%

# 1. Cost of purchasing the entire stock holding
purchase_cost = units * purchase_price_per_stock

# 2. Commission paid during purchase
purchase_commission = purchase_cost * commission_rate

# 3. Selling price of the entire stock holding
selling_value = units * selling_price_per_stock

# 4. Commission paid during sale
selling_commission = selling_value * commission_rate

# 5. Profit margin (after considering both commissions)
profit = (selling_value - selling_commission) - (purchase_cost + purchase_commission)

# Display the results
print("1. Cost of purchasing the entire stock holding: $", round(purchase_cost, 2))
print("2. Commission paid during purchase: $", round(purchase_commission, 2))
print("3. Selling price of the entire stock holding: $", round(selling_value, 2))
print("4. Commission paid during sale: $", round(selling_commission, 2))
print("5. Profit margin (after all commissions): $", round(profit, 2))
```

Write a Python function named update_inventory  that  takes  the  current  product_inventory  and  a  dictionary representing  the  new  shipment  with  product  names  and  quantities.  The  function should update the inventory with the new shipment quantities. (6 Marks)

iii)You want to offer discounts on certain products. Write a Python function named apply_discount  that  takes  the  current  product_inventoryand  a  dictionary representing  the  discounted  products  with  their  new  prices.  The  function  should update the inventory with the discounted prices. (7 Marks)

```python
# i) Create a dictionary to store product names and their quantities
product_inventory = {
    "Laptop": 10,
    "Smartphone": 25,
    "Headphones": 50,
    "Keyboard": 15,
    "Mouse": 20
}
```

```python
# ii) Function to update inventory with new shipment quantities
def update_inventory(current_inventory, new_shipment):
    for product, quantity in new_shipment.items():
        # If product already exists, add to the existing quantity
        if product in current_inventory:
            current_inventory[product] += quantity
        else:
            # Add new product to inventory
            current_inventory[product] = quantity
    return current_inventory
```

```python
new_shipment = {
    "Laptop": 5,
    "Smartwatch": 10,
    "Mouse": 8
}

product_inventory = update_inventory(product_inventory, new_shipment)
print("Updated Inventory:", product_inventory)
```

```python
# Modified inventory with quantity and price
product_inventory = {
    "Laptop": {"quantity": 10, "price": 80000},
    "Smartphone": {"quantity": 25, "price": 30000},
    "Headphones": {"quantity": 50, "price": 2000},
    "Keyboard": {"quantity": 15, "price": 1500},
    "Mouse": {"quantity": 20, "price": 800}
}

# iii) Function to apply discounts on certain products
def apply_discount(current_inventory, discounted_products):
    # Update the inventory with discounted prices
    # current_inventory.update(discounted_products)
    for product, new_price in discounted_products.items():
        if product in current_inventory:
            current_inventory[product]["price"] = new_price
    return current_inventory

discounted_products = {
    "Laptop": 75000,
    "Headphones": 1800
}

product_inventory = apply_discount(product_inventory, discounted_products)
print("Inventory after applying discounts:", product_inventory)
```
