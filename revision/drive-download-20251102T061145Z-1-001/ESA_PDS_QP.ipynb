### 3e.). Fibonacci Series is computed based on the following formula:                                          (5 marks)

f(n)=0 if n=0

f(n)=1 if n=1

f(n)=f(n-1)+f(n-2) if n>1

Write a program using list comprehension to print the Fibonacci Sequence in the comma-separated form with a given n input by 

user input.

Sample input: 5

Output:  	0,1,1,2,


```python
fib = lambda n: [0, 1] if n == 0 else fib(n - 1) + [fib(n - 1)[-1] + fib(n - 1)[-2]]
print(fib(5))
```

```python
def fibSeries(n):
  a, b = 0, 1
  res = []
  for _ in range(n):
    res.append(a)
    a, b = b, a + b
  return res

fibSeries(5)
```

```python
def fib(n):
  a, b = 0, 1
  while a < n:
    print(a, end=' ')
    a, b = b, a + b
```

```python
fib(5)
```

***
#### Sample Input/Output:

Enter a number between 1 to 15: 3

Enter a number between 1 to 15: 7

Enter a number between 1 to 15: 12

Enter a number between 1 to 15: 13

Enter a number between 1 to 15: 4

Enter a number between 1 to 15: 5

Output list :  [3, 10, 7, 10, 10, 4, 5, 6, 10, 10]

***

```python
res = []
for i in range(10):
    n = input('Enter a number between 1 and 15: ')
    res.append(n)

final = list(map(lambda x: 10 if int(x) > 10 else int(x), res))
print(final)
```
